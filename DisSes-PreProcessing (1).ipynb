Gerekli kütüphaneleri yükleyerek başlıyoruz

```python
# Import Numpy and Matplotlib
import numpy
import matplotlib.pyplot

import os
```

Bilgisayarda kayıtlı olan resimleri yüklüyoruz

```python
X = []
Y = []

path = "Spectograms/"

for i in range(10):
    fold = path + str(i)
    curCatImgs = os.listdir(fold)
    
    for img in curCatImgs:
        curImg = matplotlib.image.imread(fold + "/" + img)
        ImgCut = numpy.delete(curImg, 3, 2)
        X.append(ImgCut)
        Y.append(i)
```

Elde edilen veri setlerini böl

```python
x_train = numpy.array(X[:6986])
y_train = numpy.array(Y[:6986])

x_temp = X[6986:]
y_temp = Y[6986:]

x_test = numpy.array(x_temp[873:])
y_test = numpy.array(y_temp[873:])

x_valid = numpy.array(x_temp[:873])
y_valid = numpy.array(y_temp[:873])
```
